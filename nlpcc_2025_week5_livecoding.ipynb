```python
# Step 1: Install the OpenAI library
!pip install openai -q

# Step 2: Import necessary libraries
import openai
from google.colab import userdata

# Step 3: Retrieve the API key from Colab secrets
try:
    api_key = userdata.get('openai_key')
    if not api_key:
        raise ValueError("API key 'openai_key' not found in Colab secrets. Please add it.")
except userdata.SecretNotFoundError:
    print("🚨 Secret 'openai_key' not found. Please follow these steps:")
    print("1. Click the key icon (🔑) in the left sidebar.")
    print("2. Click 'ADD NEW SECRET'.")
    print("3. Name: openai_key, Value: your_actual_api_key_here")
    print("4. Ensure 'Notebook access' is toggled ON.")
    raise # Re-raise the exception to stop execution

# Step 4: Initialize the OpenAI client
client = openai.OpenAI(api_key=api_key)

# Step 5: Define the chat function
def get_chat_response(user_input, model_name="gpt-4o"): # Using gpt-4o as a reliable default
    """
    Sends a message to the OpenAI API and gets a response.
    """
    print(f"\n🤖 Using model: {model_name}")
    try:
        response = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": user_input}
            ]
        )
        assistant_reply = response.choices[0].message.content
        return assistant_reply
    except openai.APIError as e:
        print(f"An API error occurred: {e}")
        if "gpt-4.1" in str(e) or "does not exist" in str(e).lower():
            print("Error might be due to the model 'gpt-4.1' not being available to your key or not existing.")
            print("Consider trying 'gpt-4o', 'gpt-4-turbo', or 'gpt-4'.")
        return "Sorry, I encountered an error."
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return "Sorry, I encountered an unexpected error."

# Step 6: Get user input and display the response
if __name__ == "__main__":
    print("OpenAI Chatbot Initialized.")
    print("Type 'exit' or 'quit' to end the chat.")

    # --- CHOOSE YOUR MODEL ---
    # If "gpt-4.1" is truly available to you, use it here.
    # Otherwise, stick with a known model like "gpt-4o".
    chosen_model = "gpt-4.1" # As per your request
    # chosen_model = "gpt-4o" # A reliable alternative

    while True:
        user_message = input("\nYou (type your message): ")
        if user_message.lower() in ['exit', 'quit']:
            print("Exiting chat. Goodbye!")
            break

        if not user_message.strip():
            print("AI: Please enter a message.")
            continue

        ai_response = get_chat_response(user_message, model_name=chosen_model)
        print(f"AI: {ai_response}")
```

```python
# Step 1: Install necessary libraries
!pip install openai pandas -q

# Step 2: Import libraries
import openai
import pandas as pd
from google.colab import files, userdata
import io # For handling file uploads

# Step 3: Retrieve the API key from Colab secrets
try:
    api_key = userdata.get('openai_key')
    if not api_key:
        raise ValueError("API key 'openai_key' not found in Colab secrets. Please add it.")
except userdata.SecretNotFoundError:
    print("🚨 Secret 'openai_key' not found. Please follow these steps:")
    print("1. Click the key icon (🔑) in the left sidebar.")
    print("2. Click 'ADD NEW SECRET'.")
    print("3. Name: openai_key, Value: your_actual_api_key_here")
    print("4. Ensure 'Notebook access' is toggled ON.")
    raise
except Exception as e:
    print(f"An error occurred while retrieving the API key: {e}")
    raise

# Step 4: Initialize the OpenAI client
client = openai.OpenAI(api_key=api_key)

# Step 5: Function to get summary from LLM
def get_summary_from_llm(text_to_summarize, user_prompt, model_name="gpt-4o"): # Using a cost-effective model for summarization
    """
    Sends text and a custom prompt to the OpenAI API for summarization.
    """
    if not text_to_summarize or pd.isna(text_to_summarize): # Handle empty or NaN values
        return "Input text was empty or missing."

    # Constructing the messages for the LLM
    # The user_prompt guides how the text_to_summarize should be processed.
    messages = [
        {"role": "system", "content": "You are an expert at summarizing text based on user instructions."},
        {"role": "user", "content": f"{user_prompt}\n\nHere is the text to process:\n\"\"\"\n{text_to_summarize}\n\"\"\""}
    ]

    try:
        print(f"   Sending to model ({model_name}) for summarization...")
        response = client.chat.completions.create(
            model=model_name,
            messages=messages
        )
        summary = response.choices[0].message.content
        return summary
    except openai.APIError as e:
        print(f"   ❌ API Error for text: '{str(text_to_summarize)[:50]}...': {e}")
        return f"API Error: {e}"
    except Exception as e:
        print(f"   ❌ Unexpected Error for text: '{str(text_to_summarize)[:50]}...': {e}")
        return f"Unexpected Error: {e}"

# Step 6: Main processing logic
if __name__ == "__main__":
    print("--- CSV Text Summarization with OpenAI LLM ---")

    # --- CHOOSE YOUR MODEL ---
    # For summarization, gpt-3.5-turbo is often sufficient and more cost-effective.
    # If you need higher quality or have access to "gpt-4.1" and prefer it:
    # chosen_model = "gpt-4.1"
    chosen_model = "gpt-3.5-turbo" # Recommended for general summarization
    # chosen_model = "gpt-4o" # Another good option

    print(f"\n🤖 Using LLM model: {chosen_model}")

    # 1. Upload CSV
    print("\n1. Please upload your CSV file:")
    uploaded = files.upload()

    if not uploaded:
        print("No file uploaded. Exiting.")
    else:
        file_name = list(uploaded.keys())[0]
        print(f"\nSuccessfully uploaded '{file_name}'")

        try:
            # Read the CSV into a pandas DataFrame
            # Using io.BytesIO to handle the uploaded file content
            df = pd.read_csv(io.BytesIO(uploaded[file_name]))
            print("\nDataFrame loaded successfully. Here are the first 5 rows:")
            print(df.head())
            print("\nDataFrame columns:", df.columns.tolist())

            # 2. Get column name for summarization
            while True:
                column_to_summarize = input("\n2. Enter the name of the column containing the text to summarize: ").strip()
                if column_to_summarize in df.columns:
                    break
                else:
                    print(f"Error: Column '{column_to_summarize}' not found in the DataFrame. Please choose from: {df.columns.tolist()}")

            # 3. Get user's summarization prompt
            user_summarization_prompt = input("\n3. Enter your prompt for summarization (e.g., 'Summarize this text in one sentence', 'Extract key entities', 'What is the main topic?'):\n").strip()
            if not user_summarization_prompt:
                print("No summarization prompt provided. Using a default: 'Summarize the following text.'")
                user_summarization_prompt = "Summarize the following text."

            print(f"\nOkay, I will process the text in the '{column_to_summarize}' column.")
            print(f"Using your prompt: '{user_summarization_prompt}'")
            print("\nProcessing... This may take a while depending on the number of rows and text length.")

            # 4. Process each row and get summaries
            llm_responses = []
            total_rows = len(df)
            for index, row in df.iterrows():
                text = str(row[column_to_summarize]) # Ensure text is string
                print(f"\nProcessing row {index + 1}/{total_rows}...")
                summary = get_summary_from_llm(text, user_summarization_prompt, model_name=chosen_model)
                llm_responses.append(summary)
                # Optional: print immediate result for long processes
                # print(f"   Summary for row {index + 1}: {summary[:100]}...")


            # 5. Add summaries as a new column
            df['response'] = llm_responses
            print("\n\n✅ Processing complete!")
            print("The 'response' column has been added to your DataFrame.")
            print("\nUpdated DataFrame (first 5 rows with the new 'response' column):")
            print(df[['_id',column_to_summarize, 'response']].head() if '_id' in df.columns else df[[column_to_summarize, 'response']].head()) # Show relevant columns

            print("\nFull updated DataFrame (first 20 rows):")
            print(df.head(20))

            # Optional: Save the updated DataFrame to a new CSV
            # output_filename = f"processed_{file_name}"
            # df.to_csv(output_filename, index=False)
            # print(f"\nUpdated DataFrame saved to '{output_filename}'")
            # files.download(output_filename)


        except pd.errors.EmptyDataError:
            print("Error: The uploaded CSV file is empty.")
        except Exception as e:
            print(f"An error occurred during processing: {e}")
```

```python
# prompt: can we make 2 new columns, food detected and sentiment, based on the "response" column?
# {
#     "food_detected": [],
#     "sentiment": "positive"
# }

def process_response(response_text):
    """
    Parses the response string to extract food detected and sentiment.
    Assumes the response is a string representation of a dictionary
    like '{"food_detected": ["pizza", "pasta"], "sentiment": "positive"}'
    or can handle other text if the dictionary is not strictly present.
    """
    food_detected = []
    sentiment = "neutral" # Default sentiment

    if isinstance(response_text, str):
        try:
            # Attempt to parse as JSON/dictionary string
            import json
            response_dict = json.loads(response_text)
            if isinstance(response_dict, dict):
                if "food_detected" in response_dict and isinstance(response_dict["food_detected"], list):
                    food_detected = response_dict["food_detected"]
                if "sentiment" in response_dict and isinstance(response_dict["sentiment"], str):
                    sentiment = response_dict["sentiment"].lower()
        except json.JSONDecodeError:
            # If not a valid JSON string, try simple keyword matching for sentiment
            if "positive" in response_text.lower():
                sentiment = "positive"
            elif "negative" in response_text.lower():
                sentiment = "negative"
            # Food detection is harder without structure, might need another LLM call
            # For this function, we'll return empty list if JSON fails
            pass
        except Exception as e:
            print(f"Error processing response string: {response_text[:100]}... Error: {e}")

    return food_detected, sentiment

# Apply the function to the 'response' column to create new columns
# Using .apply() is common, but can be slow for very large DataFrames.
# List comprehensions are often faster.

food_and_sentiment_results = [process_response(resp) for resp in df['response']]

# Unpack the results into two lists
df['food_detected'] = [item[0] for item in food_and_sentiment_results]
df['sentiment'] = [item[1] for item in food_and_sentiment_results]


# Display the updated DataFrame with the new columns
print("\nDataFrame with 'food_detected' and 'sentiment' columns added:")
print(df[['response', 'food_detected', 'sentiment']].head())
```
